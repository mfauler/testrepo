### The following code displays a standard normal distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

mu, sig = 0, 1  # mean=0, sigma=1
sns.kdeplot(random.normal(mu, sig, size=1000))  # density (pdf) plot
plt.show
```

---
The function `random.normal()` from the _numpy_ package generates `size=1000` random numbers from a normal distribution with parameter values<br>

| $$\mu$$ | $$\sigma$$ |
| :--- |:---|   
| 0 | 1 | 

The following parameters of `random.normal()` are used:
- **loc**: mean ($\mu$)
- **scale**: standard deviation ($\sigma$)
- **size**: number of random values
